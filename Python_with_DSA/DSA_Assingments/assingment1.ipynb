```python
# Part 1: Variables and Memory Allocation
# Exercise 1.1: Memory Exploration

# 1. Creating variables of different types
number = 24                 # Integer (immutable)
decimal = 12.34             # Float (immutable)
char = "my name is shivam"  # String (immutable)
my_list = [12, 23, 34, 45, 56, 67, "shivam"]  # List (mutable)
my_tuple = (1, 2, 3, 4, 3, 2, 1)              # Tuple (immutable)
```

```python
# 2. Printing the memory address (ID) of each variable
print("Memory address of int:    ", id(number))
print("Memory address of float:  ", id(decimal))
print("Memory address of string: ", id(char))
print("Memory address of list:   ", id(my_list))
print("Memory address of tuple:  ", id(my_tuple))
```

```python
# 3. Creating a second variable that references the same object
number2 = number           # same object as number
decimal2 = decimal
char2 = char
list2 = my_list            # list2 points to the same list object
tuple2 = my_tuple

# Display the new variables
print("Reference variables:")
print("number2:", number2, "| decimal2:", decimal2, "| char2:", char2)
print("list2:", list2)
print("tuple2:", tuple2)
```

```python

# 4. Modifying original variables

# IMMUTABLE EXAMPLE (int)
x = 10
y = x  # y points to the same int object
print("Before modifying x:")
print(f"x = {x}, y = {y}")
x += 10  # x now points to a new int object (20)
print("After modifying x:")
print(f"x = {x}, y = {y}")
# y remains unchanged because int is immutable

# MUTABLE EXAMPLE (list)
my_list2 = [12, 23, 34, 45]
my_list3 = my_list2  # both variables point to the same list object
print("Before modifying the list:")
print("my_list2:", my_list2)
print("my_list3:", my_list3)

# Modify the original list
my_list2.append(20)
print("After appending 20 to my_list2:")
print("my_list2:", my_list2)
print("my_list3:", my_list3)
# my_list3 also shows the change because lists are mutable

# To avoid this, use copy()
my_list4 = [1, 2, 3, 4]
my_list5 = my_list4.copy()  # creates a shallow copy (new list)
print("Using copy() to avoid shared reference:")
print("Before modifying my_list4:")
print("my_list4:", my_list4)
print("my_list5:", my_list5)

# Now modify my_list4
my_list4.append(5)
print("After modifying my_list4:")
print("my_list4:", my_list4)
print("my_list5:", my_list5)
# my_list5 remains unchanged because it's a separate object
```

```python
#Define global variables accessible throughout the script.
var=12                           #global variable
s="shivam"                       #global variable

print(f"Global variables var:{var} and s:{s}")
print("id of var:",id(var),"id of s:",id(s))

def without_modify():
    var=24          # local variable
    s="Sharma"      #local variable
    print("inside without_modify()")
    print("local var=",var ,"string=",s)
    print("id of var:",id(var),"id of s:",id(s))
    print("these are the local variables which do not affect the global one")

def with_modify():
    global var
    var=48  # this modify the global var
    print("inside with_modify()")
    print(f"global var after update :{var}, id of global var :{id(var)}")
    print(f"'global var' affect the global varibale")

print("\n calling without modify()")
without_modify()
print("\n calling with_modify()")
with_modify()
```

```python
# Creates at least one variable of each of these types: int, float, complex, bool, str, and None.
a = 23                          # int
b = 23.34                       # float
c = 3 + 6j                      # complex
d = True                        # bool
e = "my name is shivam"         # str
f = None                        # NoneType

# Checking the data type of each variable
print("Checking the data type of each variable:")
print(type(a))  # int
print(type(b))  # float
print(type(c))  # complex
print(type(d))  # bool
print(type(e))  # str
print(type(f))  # NoneType

# Using isinstance() to check the variable are of specific type
result = isinstance(a, int)
result1 = isinstance(b, float)
result2 = isinstance(c, complex)
result3 = isinstance(d, bool)
result4 = isinstance(e, str)

# You can't use isinstance(f, None) because None is not a class; use 'type(f) is NoneType' or custom check
def check_none(x):
    if x is None:
        return True
    else:
        return False

result5 = check_none(f)

print("\nUses isinstance() to check if variables are of specific types:")
print("Variable 'a' is int:", result)
print("Variable 'b' is float:", result1)
print("Variable 'c' is complex:", result2)
print("Variable 'd' is bool:", result3)
print("Variable 'e' is string:", result4)
print("Variable 'f' is NoneType:", result5)

# Python automatically converts data types during operations
eg1 = True + 1       # bool + int = int
eg2 = 12 + 12.34     # int + float = float
eg3 = 12 / 23        # int / int = float

print("\nThree examples where Python automatically converts types in expressions:")
print("True + 1 (bool + int) →", eg1, "| Type:", type(eg1))  # int
print("12 + 12.34 (int + float) →", eg2, "| Type:", type(eg2))  # float
print("12 / 23 (int / int) →", eg3, "| Type:", type(eg3))  # float
```

```python
# Converts decimal to binary 
def decimal_to_binary(n):
    if n==0:
        return "0"
    binary=""
    while n>0:
        reminder=n%2
        binary=str(reminder)+binary
        n=n//2
    return binary

# Converts decimal to octal
def decimal_to_oct(n):
    if n==0:
        return "0"
    oct=""
    while n>0:
        reminder=n%8
        oct=str(reminder)+oct
        n=n//8
    return oct

# Converts decimal to hexdecimal
def decimal_to_hexadecimal(n):
    if n==0:
        return "0"
    hex_chars = "0123456789ABCDEF"
    hexadecimal = ""
    while n > 0:
        remainder = n % 16
        hexadecimal = hex_chars[remainder] + hexadecimal
        n = n // 16
    return hexadecimal


# Converts decimal to binary,octal,hexadecimal
def main():
    user_input=int(input("enter a number"))
    print("you have enter :",user_input)
    print("binary:",decimal_to_binary(user_input))
    print("octal:",decimal_to_oct(user_input))
    print("hexadecimal:",decimal_to_hexadecimal(user_input))

#convert binary to decimal
def binary_to_decimal(a_str):
    decimal=0
    for i in digit,enumerate(reversed(a_str)):
        decimal+=int(digit)*(2**i)
    return decimal

# Converts octal string to decimal
def octal_to_decimal(o_str):
    decimal = 0
    for i, digit in enumerate(reversed(o_str)):
        decimal += int(digit) * (8 ** i)
    return decimal

# Converts hexadecimal string to decimal
def hexadecimal_to_decimal(h_str):
    hex_chars = "0123456789ABCDEF"
    h_str = h_str.upper()
    decimal = 0
    for i, digit in enumerate(reversed(h_str)):
        decimal += hex_chars.index(digit) * (16 ** i)
    return decimal


test_numbers = [0, 7, 15, 58, 255]

for num in test_numbers:
    print(f"Number: {num}")
    
    # Our custom functions
    binary_custom = decimal_to_binary(num)
    octal_custom = decimal_to_oct(num)
    hex_custom = decimal_to_hexadecimal(num)

    # Python built-in functions 
    binary_builtin = bin(num)[2:]      
    octal_builtin = oct(num)[2:]      
    hex_builtin = hex(num)[2:].upper()

    print(f"  Binary     : Custom: {binary_custom} | Built-in: {binary_builtin}")
    print(f"  Octal      : Custom: {octal_custom} | Built-in: {octal_builtin}")
    print(f"  Hexadecimal: Custom: {hex_custom} | Built-in: {hex_builtin}")
```

```python
import math

#1 math.sqrt() - Square Root
def example_sqrt():
    number = 25
    result = math.sqrt(number)
    print(f"Square root of {number} is: {result}")

#2 math.factorial() - Factorial
def example_factorial():
    number = 5
    result = math.factorial(number)
    print(f"Factorial of {number} is: {result}")

#3: math.pow() - Power
def example_pow():
    base = 2
    exponent = 3
    result = math.pow(base, exponent)
    print(f"{base} raised to the power of {exponent} is: {result}")

#4: math.log() - Natural Logarithm
def example_log():
    number = 10
    result = math.log(number)
    print(f"Natural logarithm of {number} is: {result}")

#5: math.log10() - Logarithm Base 10
def example_log10():
    number = 1000
    result = math.log10(number)
    print(f"Base 10 logarithm of {number} is: {result}")

#6: math.sin() - Sine of an Angle (radians)
def example_sin():
    angle = math.radians(30)  # Converting degrees to radians
    result = math.sin(angle)
    print(f"Sine of 30 degrees is: {result}")

#7: math.cos() - Cosine of an Angle (radians)
def example_cos():
    angle = math.radians(60)  # Converting degrees to radians
    result = math.cos(angle)
    print(f"Cosine of 60 degrees is: {result}")

#8: math.tan() - Tangent of an Angle (radians)
def example_tan():
    angle = math.radians(45)  # Converting degrees to radians
    result = math.tan(angle)
    print(f"Tangent of 45 degrees is: {result}")

#9: math.degrees() - Convert radians to degrees
def example_degrees():
    radians = math.pi / 4  # 45 degrees in radians
    result = math.degrees(radians)
    print(f"{radians} radians is equal to {result} degrees")

#10: math.ceil() - Ceiling of a Number
def example_ceil():
    number = 5.3
    result = math.ceil(number)
    print(f"Ceiling of {number} is: {result}")

#11: math.floor() - Floor of a Number
def example_floor():
    number = 5.8
    result = math.floor(number)
    print(f"Floor of {number} is: {result}")

# Function 1: Roots of a quadratic equation
def quadratic_roots(a, b, c):
    discriminant = math.pow(b, 2) - 4 * a * c
    if discriminant < 0:
        print("No real roots")
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2 * a)
        root2 = (-b - math.sqrt(discriminant)) / (2 * a)
        print(f"Roots of the quadratic equation {a}x^2 + {b}x + {c} = 0 are: {root1} and {root2}")

# Function 2: Convert between degrees and radians
def convert_angle(value, unit='radian'):
    if unit == 'radian':
        result = math.degrees(value)
        print(f"{value} radians is {result} degrees")
    elif unit == 'degree':
        result = math.radians(value)
        print(f"{value} degrees is {result} radians")
    else:
        print("Invalid unit specified. Use 'degree' or 'radian'.")
```

```python
def main():
    # Call all examples
    example_sqrt()
    example_factorial()
    example_pow()
    example_log()
    example_log10()
    example_sin()
    example_cos()
    example_tan()
    example_degrees()
    example_ceil()
    example_floor()
    
    # Calculate roots of a quadratic equation (e.g., x^2 - 3x + 2 = 0)
    quadratic_roots(1, -3, 2)
    
    # Angle conversions (example)
    convert_angle(45, 'degree')
    convert_angle(math.pi / 4, 'radian')

    
main()
```
